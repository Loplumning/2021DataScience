```python
import pandas as pd
```

```python
SVM = pd.read_csv('upload_result/SVM.csv')
DNN = pd.read_csv('upload_result/DNN.csv')
XGB = pd.read_csv('upload_result/XGBoost.csv')
CatB = pd.read_csv('upload_result/CatBoost.csv')
RanFor = pd.read_csv('upload_result/RandomForest.csv')
```

```python
vote = []
all_0_1 = 0
for i in range(2000):
    num_of_0 = sum([SVM.Exited[i]==0, DNN.Exited[i]==0, XGB.Exited[i]==0, CatB.Exited[i]==0, RanFor.Exited[i]==0])
    if num_of_0 >= 3:
        vote.append(0)
    else:
        vote.append(1)
    if num_of_0==0 | num_of_0==5:
        all_0_1 += 1
```

```python
upload_data = pd.DataFrame({'RowNumber': SVM.RowNumber, 'Exited':vote})
```
