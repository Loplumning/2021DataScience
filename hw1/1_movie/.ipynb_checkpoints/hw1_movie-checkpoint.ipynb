### Q2 : The actor generating the highest average revenue?


```python
###  create dataframe of every single actor  ###
df_rv = pd.DataFrame()

for i in range(len(df)) :
    temp = df['Actors'][i]  
    temp = temp.split('|')
    act = []
    title = []
    revenue = []
    for j in range(len(temp)) :
        tempj = temp[j].strip()
        tempj = tempj.replace('?', '')
        act.append(tempj)
        title.append(df['Title'][i])
        revenue.append(df['Revenue (Millions)'][i])

    tempdf = pd.DataFrame({'actor' : act, 'movie' : title, 'revenue' : revenue})
    df_rv = df_rv.append(tempdf)

###  sort by actor, reset index  ###
df_rv = df_rv.sort_values(by='actor')
df_rv = df_rv.reset_index()
df_rv = df_rv.drop(labels='index', axis='columns')


###  calculate the average  ###
df_rv_avg = df_rv.groupby('actor').mean()
df_rv_avg = df_rv_avg.sort_values(by='revenue', ascending=False)
df_rv_avg[df_rv_avg['revenue'] == df_rv_avg['revenue'].max()]
```

##### answer : 'Daisy Ridley' and 'John Boyega'

### Q4 : Top‐3 directors who collaborate with the most actors? 

```python
###  create dataframe of every single actor with director  ###
df_dir = pd.DataFrame()

for i in range(len(df)) :
    temp = df['Actors'][i]  
    temp = temp.split('|')
    act = []
    title = []
    director = []
    for j in range(len(temp)) :
        tempj = temp[j].strip()
        tempj = tempj.replace('?', '')
        act.append(tempj)
        title.append(df['Title'][i])
        director.append(df['Director'][i])

    tempdf = pd.DataFrame({'actor' : act, 'movie' : title, 'director' : director})
    df_dir = df_dir.append(tempdf)

###  sort by director+actor, reset index  ###
df_dir = df_dir.sort_values(by=['director', 'actor'])
df_dir = df_dir.reset_index()
df_dir = df_dir.drop(labels='index', axis='columns')

###  calculate the count  ###
df_dir_count = pd.DataFrame()

x = df_dir['director']
dir_list = sorted(set(x))

director = []
co_actor = []
count = []

for i in range(len(dir_list)) :
    a = df_dir['actor'][df_dir['director'] == dir_list[i]]
    b = sorted(set(a))
    director.append(dir_list[i])
    co_actor.append(b)
    count.append(len(b))

df_dir_count = pd.DataFrame({'director' : director, 'collaborated actors' : co_actor, 'count' : count})

###  result  ###
df_dir_count.sort_values(by='count', ascending=False).head()
```

##### answer : 'Ridley Scott', 'M. Night Shyamalan', 'Paul W.S. Anderson' and 'Danny Boyle'

### Q5 : Top‐2 actors playing in the most genres of movies? 

```python
###  create dataframe of every single actor with genre  ###
df_0 = pd.DataFrame()

for i in range(len(df)) : 
    temp = df['Actors'][i]  
    temp = temp.split('|')
    temp_g = df['Genre'][i]
    temp_g = temp_g.split('|')
    act = []
    title = []
    genre = []
    for j in range(len(temp)) :
        tempj = temp[j].strip()
        tempj = tempj.replace('?', '')
        
        for k in range(len(temp_g)) :
            act.append(tempj)
            title.append(df['Title'][i])
            genre.append(temp_g[k]) 

    tempdf = pd.DataFrame({'actor' : act, 'genre' : genre, 'movie' : title})
    df_0 = df_0.append(tempdf)

    
###  sort by actor, reset index  ###
df_0 = df_0.sort_values(by=['actor', 'genre'])
df_0 = df_0.reset_index()
df_0 = df_0.drop(labels='index', axis='columns')


###   calculate the count   ###
df_gcount = pd.DataFrame()

x = df_0['actor']
act_list = sorted(set(x))

actor = []
genre_list = []
count = []

for i in range(len(act_list)) :
    a = df_0['genre'][df_0['actor'] == act_list[i]]
    b = sorted(set(a))
    actor.append(act_list[i])
    genre_list.append(b)
    count.append(len(b))

df_gcount = pd.DataFrame({'actor' : actor, 'genre of movie' : genre_list, 'count' : count})


###  result  ###
result = df_gcount.sort_values(by='count', ascending=False)
c = result['count']
c = sorted(set(c), reverse=True)

top1 = df_gcount[df_gcount['count'] == c[0]]
top2 = top1.append(df_gcount[df_gcount['count'] == c[1]])

top2
```

##### answer : top1 - 'Brad Pitt'

### Q6 : Top‐3 actors whose movies lead to the largest maximum gap of years? 

```python
###  create dataframe of every single actor  ###
df_0 = pd.DataFrame()

for i in range(len(df)) : 
    temp = df['Actors'][i]  
    temp = temp.split('|')
    act = []
    title = []
    year = []
    for j in range(len(temp)) :
        tempj = temp[j].strip()
        tempj = tempj.replace('?', '')
        act.append(tempj)
        title.append(df['Title'][i])
        year.append(df['Year'][i])

    tempdf = pd.DataFrame({'actor' : act, 'movie' : title, 'year' : year})
    df_0 = df_0.append(tempdf)

###  sort by actor, reset index  ###
df_0 = df_0.sort_values(by='actor')
df_0 = df_0.reset_index()
df_0 = df_0.drop(labels='index', axis='columns')


###  calculate the gap of year  ###
a = df_0.groupby('actor')['year']
result = pd.DataFrame({'gap' : a.max() - a.min()})

result = result.sort_values(by='gap', ascending=False)
result = result.reset_index()
result.head()
result['actor'][result['gap'] == 10]
```

##### answer : There are 53 actors whose largest maximum gap of years are all the same, which is, 10(years).
